#### In this notebook, we extract text from a few PDF documents and apply inverted indexing to it. The documents are stories and novels in English. By default we are reading only the first 500 pages.

Necessary Imports

```python
import nltk
import PyPDF2
nltk.download('punkt')
nltk.download('stopwords')
import glob
import pprint

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

## Read data from PDF

```python
def readPDF(path,numPages):
    Text=""
    pdfFileObj = open(path, 'rb') 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
    totalPages=pdfReader.numPages
    if(numPages>totalPages):
        numPages=totalPages
    print(totalPages)
    for i in range(numPages):
            pageObj = pdfReader.getPage(i)
            try:
                Text=Text+str(pageObj.extractText())
            except:
                print("Read error")

    pdfFileObj.close()
    return Text
    
```

## Find all PDF files in directory

```python
from nltk.tokenize import sent_tokenize, word_tokenize
text_files=glob.glob('*.pdf')
text_files
```

## Extract and Process Data
### Word Tokenization, Removing Stop words, Stemming and Indexing

```python
stop_words = set(stopwords.words('english'))
index=dict()
for file in text_files:
    index[file]=dict()
numOfPages=500
for file in text_files:
    file_content=""
    file_name=file
    try:
        file_content=readPDF(file,numOfPages)
        tokenize_content=word_tokenize(file_content)
        filtered_sentence = []

        for w in tokenize_content:
            if w not in stop_words:
                filtered_sentence.append(w)
        ps = PorterStemmer()
        stemmed_sentence=[]
        for w in filtered_sentence:
                stemmed_sentence.append(ps.stem(w))
        for w in stemmed_sentence:
            if w in index[file_name]:
                index[file_name][w]+=1
            else:
                index[file_name][w]=1
        print("Success")        
        
                
    except UnicodeDecodeError:
        print("Error reading")
      
    
```

## Index has been populated with word frequncies of each file

```python
for i in index:
    print(i)
    print(len(index[i]))
```

## Getting inverted index from index

```python
inverted_index=dict()
for file in index:
    for w in index[file]:
        if w in inverted_index:
            inverted_index[w][file]=index[file][w]
        else:
            inverted_index[w]=dict()
            inverted_index[w][file]=index[file][w]
            
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(inverted_index)
```
